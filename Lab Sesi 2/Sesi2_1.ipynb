```python
import numpy as np

# Define Array
x_arr = [
    [
        [4, -1, 2],
        [3, 6, -1],
        [2, 1, 5],
    ],
    [
        [6, -1, 2, 0],
        [3, 9, 1, -4],
        [1, 2, 7, -1],
        [2, -3, 1, 8],
    ],
    [
        [6, -1, 2, 0, 0],
        [1, 10, 3, -2, 0],
        [2, -1, 12, 1, -4],
        [0, 3, -2, 9, 1],
        [1, -2, 1, 3, 8],
    ]
]

y_arr = [
    [8, 9, 3],
    [12, 18, 14, 4],
    [14, 20, 22, 12, 15],
]
```

```python


def check_diagonal_dominant(x):
    # change into array
    x = np.array(x)
    # Take out the diagonal from the array
    diagonal = np.diag(np.abs(x))
    # Take out the non diagonal
    nondiagonal = np.sum(np.abs(x), axis=1) - diagonal
    # Check if the diagonal is greater than the non diagonal
    if np.all(diagonal >= nondiagonal):
        return True
    else:
        return False
    
def gauss_seidel(x_arr, y_arr, max_iteration, epsilon):
    # change into array
    x = np.array(x_arr)
    y = np.array(y_arr)
    # Take out the diagonal from the array
    diagonal = np.diag(x)
    # Take out the non diagonal
    x = -x
    np.fill_diagonal(x, 0)
    # Initialize the x_old
    x_old = np.zeros(len(x))
    # Start the iteration
    for i in range(max_iteration):
        x_new = np.array(x_old)
        for j,row in enumerate(x):
            # Formula of Gauss Seidel
            x_new[j] = (y[j] + np.dot(row, x_new)) / diagonal[j]
        print(f"Iteration: {i+1}, x_new : {x_new}")

        #  Error => euclidean distance
        euclidean_distance = np.sqrt(np.dot(x_old - x_new, x_old - x_new))

        if euclidean_distance < epsilon:
            print(f"Converged reached at iteration {i+1}")
            print(f"Solution: {x_new}")
            return True
        x_old = x_new
    return False

def run_gauss_seidel(x_arr, y_arr, max_iteration, epsilon):
   for i,x in enumerate(x_arr):
       if check_diagonal_dominant(x):
            print(f"System {i+1} is Diagonally Dominant")
            if gauss_seidel(x, y_arr[i], max_iteration, epsilon):
                print(f"System {i+1} is Convergent")
            else:
                print(f"System {i+1} is not Convergent")
       else:
           print(f"System {i+1} is not Diagonally Dominant")


```

```python
run_gauss_seidel(x_arr, y_arr, 100, 0.001)
```
